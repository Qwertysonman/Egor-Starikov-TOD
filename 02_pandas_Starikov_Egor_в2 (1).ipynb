```python
val = {'Количество строк': [recipes.shape[0], reviews.shape[0]], 'Количество столбцов': [recipes.shape[1], reviews.shape[1]]}
qwer = {i : [] for i in range(1, 1 + max(len(list(recipes.dtypes)), len(list(reviews.dtypes))))}
for i in range(max(len(list(recipes.dtypes)), len(list(reviews.dtypes)))):
  try:
    qwer[i+1].append(recipes.dtypes[i])
  except:
    qwer[i+1].append('NaN')
for i in range(max(len(list(recipes.dtypes)), len(list(reviews.dtypes)))):
  try:
    qwer[i+1].append(reviews.dtypes[i])
  except:
    qwer[i+1].append('NaN')
val.update(qwer)
pd.DataFrame(val, index = ['recipes', 'reviews'])
```

### Работа со строковыми данными в `pandas`

3.1  Добавьте в таблицу `recipes` столбец `description_length`, в котором хранится длина описания рецепта из столбца `description`.

```python
recipes['description_length'] = recipes['description'].apply(lambda x: len(str(x)))
recipes
```

3.2 Измените название каждого рецепта в таблице `recipes` таким образом, чтобы каждое слово в названии начиналось с прописной буквы.

```python
recipes['name'] = recipes['name'].transform(lambda x: ' '.join([i[0].upper() + i[1:] for i in x.split()]))
recipes
```

3.3 Добавьте в таблицу `recipes` столбец `name_word_count`, в котором хранится количество слов из названии рецепта (считайте, что слова в названии разделяются только пробелами). Обратите внимание, что между словами может располагаться несколько пробелов подряд.

```python
recipes['name_word_count'] = recipes['name'].apply(lambda x: len(x.split()))
recipes
```

### Объединение таблиц `pd.DataFrame`

5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. Рецепты, на которые не оставлен ни один отзыв, должны отсутствовать в полученной таблице. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и попытавшись найти строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
new_data = pd.merge(recipes, reviews, left_on = 'id', right_on = 'user_id')[['id', 'name', 'user_id', 'rating', 'review', 'recipe_id']]
coop = new_data.groupby('recipe_id', as_index = False).count()
coop.drop(['id', 'name', 'user_id', 'rating'], axis = 1, inplace = True)
coop.rename(columns = {"review": "counts"}, inplace = True)
print('Выведем количество рецептов с определенным количеством отзывов, чтобы показать какие значения попали в dataset: \n')
print(coop.groupby('counts').count())
res_dat1 = pd.merge(new_data, coop, left_on = 'recipe_id', right_on = 'recipe_id')[['id', 'name', 'user_id', 'rating', 'counts']]
res_dat1 = res_dat1[res_dat1.counts != 0]
res_dat1.drop('counts', axis = 1, inplace = True)
res_dat1
```

5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`, где столбец `review_count` содержит кол-во отзывов, оставленных на рецепт `recipe_id`. У рецептов, на которые не оставлен ни один отзыв, в столбце `review_count` должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и найдя строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
new_data = pd.merge(recipes, reviews, left_on = 'id', right_on = 'user_id')[['recipe_id', 'name', 'review']]
new_data['review_count'] = new_data.groupby('recipe_id', as_index = False).count()['review']
new_data.drop('review', axis = 1, inplace = True)
print(new_data.fillna(0.0, inplace = True))
print('Сделаем проверку на существование в dataframe значений review_count с несуществующим значением \n')
res_dat2 = new_data.groupby('review_count').count()
res_dat2
```
